```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
star_wars=pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
#star_wars.head(10)
star_wars.columns

star_wars=star_wars[pd.notnull(star_wars["RespondentID"])]
pd.notnull(star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"]).value_counts()
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
import numpy as np
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
yes_no = {"Yes": True, "No": False, np.nan:False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?",
    "Do you consider yourself to be a fan of the Star Trek franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)
star_wars[star_wars.columns[2]].value_counts()
```

**I have used the Map function to convert the columns to bool,it will be easier to plot bool values**

```python
import numpy as np
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
star_wars =star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"})
star_wars.head()
star_wars["seen_2"].value_counts()
```

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars.columns
```

**I have renamed columns for simplicity**

```python
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"})

%matplotlib inline

mean=star_wars[star_wars.columns[9:15]].mean()
movies = ["Episode I","Episode II","Episode III","Episode IV","Episode V","Episode VI"]
sns.barplot(movies,mean)
plt.title("Average Movie Ratings")
plt.xticks(rotation=45)
#plt.bar(range(1,7),me)
plt.xlabel("Movies 1-6")
plt.ylabel("Average Rating on 1-6, 1>6")
plt.show()
```

**So far I have cleaned data,more importantly columns as I have made categorial values to True or False with map function of Series**

**Lastly, I have ploted a graph of average rating for all 6 Star war Movies from 1186 responses,the lower the average ranking the better(Ranked on scale of 1-6, 1 being highest)**

```python

star_wars[star_wars.columns[3:9]].sum()
```

```python

sns.barplot(movies, star_wars[star_wars.columns[3:9]].sum())
plt.title("Response")
plt.xticks(rotation=45)
plt.xlabel("Movies 1-6")
plt.ylabel("Response")
plt.show()
```

**I calculated the Response for each film,considering the above two plot, we can see that our Response and Ratings from our survey are Related
Movie 5 and 6 have best ratings and have highest response.**


```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

sns.barplot(movies, males[males.columns[9:15]].mean())
plt.xlabel("Movies 1-6")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.title("Average Rating Male")
plt.show()

sns.barplot(movies, females[females.columns[9:15]].mean())
plt.xlabel("Movies 1-6")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.title("Average Rating Female")
plt.show()
```

```python
sns.barplot(movies, males[males.columns[3:9]].sum())
plt.xlabel("Movies 1-6")
plt.xticks(rotation=45)
plt.ylabel("Response")
plt.title("Response Male")
plt.show()

sns.barplot(movies, females[females.columns[3:9]].sum())
plt.xlabel("Movies 1-6")
plt.xticks(rotation=45)
plt.ylabel("Response")
plt.title("Response Female")
plt.show()
```

**Interestingly, more males watches episodes 1-3, but males liked them far less than females did.**

```python
BothFranchise = star_wars[(star_wars[star_wars.columns[2]]==True)&(star_wars[star_wars.columns[32]]==True)]
AB=BothFranchise.shape[0]
```

```python
Ab=star_wars[(star_wars[star_wars.columns[2]]==True)].shape[0]
aB=star_wars[(star_wars[star_wars.columns[32]]==True)].shape[0]
print(Ab,aB)
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn2_circles


# Subset sizes
s = (
    Ab,  # Ab
    aB,# aB
    AB,  # AB
)
```

```python

v = venn2(subsets=s, set_labels=('Star Wars', 'Star Trek'))

# Subset labels
v.get_label_by_id('01').set_text(' Star Trek \n427')
v.get_label_by_id('10').set_text(' Star Wars \n552')
v.get_label_by_id('11').set_text(' Both\n370')
#set Color
v.get_patch_by_id('10').set_color('purple')
v.get_patch_by_id('01').set_color('orange')
v.get_patch_by_id('11').set_color('green')

# Border styles
c = venn2_circles(subsets=s, linestyle='solid')
c[0].set_ls('dashed')  # Line style
c[0].set_lw(2.0)       # Line width
plt.show()
```
